```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#purchase_data.columns #needed to see column names
#Create variable to count unique screen names as number of players
num_players = len(purchase_data["SN"].unique())
#create dataframe
player_summary_df = pd.DataFrame({"Total Number of Players": [num_players]})
player_summary_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Set up variables for the data frame
num_items = len(purchase_data["Item ID"].unique())
count_purchID = len(purchase_data["Purchase ID"].unique())
avg_purch_price = sum(purchase_data["Price"])/count_purchID
tot_revenue = sum(purchase_data["Price"])

#create dataframe
Purch_Analysis_df = pd.DataFrame({"Number of Unique Items": [num_items],
                                  "Average Purchase Price": [avg_purch_price],
                                  "Total Number of Purchases": [count_purchID],
                                  "Total Revenue": [tot_revenue]})
#format
Purch_Analysis_df["Average Purchase Price"] = Purch_Analysis_df["Average Purchase Price"].map("${:,.2f}".format)
Purch_Analysis_df["Total Revenue"] = Purch_Analysis_df["Total Revenue"].map("${:,.2f}".format)
#display
Purch_Analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#set up variables for count and percentage of total by gender
male_players = len(purchase_data.loc[purchase_data["Gender"]=="Male"]["SN"].unique())
permale_players = male_players/num_players

female_players = len(purchase_data.loc[purchase_data["Gender"]=="Female"]["SN"].unique())
perfemale_players = female_players/num_players

other_players = len(purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed"]["SN"].unique())
perother_players = other_players/num_players 
#set up list of dictionaries
Gender_demo = [{"Gender": "Male", "Total Count": male_players, "Percentage of Players": permale_players},
                {"Gender": "Female", "Total Count": female_players, "Percentage of Players": perfemale_players},
                {"Gender": "Other / Non-Disclosed", "Total Count": other_players, "Percentage of Players": perother_players}]
#Create dataframe
Gender_df = pd.DataFrame(Gender_demo)
#format
Gender_df["Percentage of Players"] = Gender_df["Percentage of Players"].map("{:.2%}".format)
#display
Gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create needed variables
PurchBy_count_male= len(purchase_data.loc[purchase_data["Gender"]=="Male"]["Purchase ID"])
PurchBy_tot_male = purchase_data.loc[purchase_data["Gender"]=="Male"]["Price"].sum()
PurchBy_avgprice_male = PurchBy_tot_male/PurchBy_count_male
PurchBy_avgcostperperson_male = PurchBy_tot_male/male_players

PurchBy_count_female= len(purchase_data.loc[purchase_data["Gender"]=="Female"]["Purchase ID"])
PurchBy_tot_female = sum(purchase_data.loc[purchase_data["Gender"]=="Female"]["Price"])
PurchBy_avgprice_female = PurchBy_tot_female/PurchBy_count_female
PurchBy_avgcostperperson_female = PurchBy_tot_female/female_players

PurchBy_count_other= len(purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed"]["Purchase ID"])
PurchBy_tot_other = sum(purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed"]["Price"])
PurchBy_avgprice_other = PurchBy_tot_other/PurchBy_count_other
PurchBy_avgcostperperson_other = PurchBy_tot_other/other_players

#set up list of dictionaries
PurchByGender = [{"Gender": "Male", "Purchase Count": PurchBy_count_male, "Average Purchase Price": PurchBy_avgprice_male, "Total Purchase Value": PurchBy_tot_male, "Avg. Total Purchase": PurchBy_avgcostperperson_male},
                {"Gender": "Female", "Purchase Count": PurchBy_count_female, "Average Purchase Price": PurchBy_avgprice_female, "Total Purchase Value": PurchBy_tot_female, "Avg. Total Purchase": PurchBy_avgcostperperson_female},
                {"Gender": "Other / Non-Disclosed", "Purchase Count": PurchBy_count_other, "Average Purchase Price": PurchBy_avgprice_other, "Total Purchase Value": PurchBy_tot_other, "Avg. Total Purchase": PurchBy_avgcostperperson_other}]
#create dataframe
PurchByGender_df = pd.DataFrame(PurchByGender)

#format and display
PurchByGender_df["Average Purchase Price"] = PurchByGender_df["Average Purchase Price"].map("${:,.2f}".format)
PurchByGender_df["Total Purchase Value"] = PurchByGender_df["Total Purchase Value"].map("${:,.2f}".format)
PurchByGender_df["Avg. Total Purchase"] = PurchByGender_df["Avg. Total Purchase"].map("${:,.2f}".format)
PurchByGender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Since dealing with player counts, need a way to identify unique players and their ages. (Remove dup SN.)
age_df = purchase_data.drop_duplicates(['SN']) #drop people with more than one charge
#Bins are implied with expected output. Start with 0 and end of bin is 1 less than start of next bin since integers
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200] 
#Group names are given.
group_names = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+"]
#Bin list must be one longer than group names, add arbitray upper limit?
#print(len(bins)) #must be one more than group_names
#print(len(group_names))
#create a "cut" and add the new age categories to the dataframe. Check that this is working.
age_df["Age Category"] = pd.cut(age_df["Age"], bins, labels = group_names, include_lowest=True)

#group by
age_df = age_df.groupby("Age Category")
#Create dictionary of lists (change in method from above setup. Practicing both ways.)
age_distro = {"Total Count": age_df["SN"].count(),
              "Percentage of Players": age_df["SN"].count()/age_df["SN"].count().sum()}
#Create dataframe
age_distro_df = pd.DataFrame(age_distro)
#format (only one value requires)
age_distro_df["Percentage of Players"] = age_distro_df["Percentage of Players"].map("{:.2%}".format)
#display
age_distro_df
```

## Purchasing Analysis (Age)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#use original dataframe
age_purch = purchase_data
#same binning as the Age Demo above. But this time we are analyzing transactions, not players.
#duplicate screen names are ok.
age_purch["Age Category"] = pd.cut(age_purch["Age"], bins, labels = group_names, include_lowest=True)
age_purch = age_purch.groupby("Age Category")

#set up dictionary of lists
age_purch_df = {"Purchase Count": age_purch["Purchase ID"].count(),
             "Average Purchase Price": age_purch["Price"].sum()/age_purch["Purchase ID"].count(),
             "Total Purchase Value": age_purch["Price"].sum(), 
             "Avg. Total Purchase per Person": age_purch["Price"].sum()/age_df["SN"].count()}
#format experiment: testing whether formatting before creating dataframe has impact 
age_purch_df["Average Purchase Price"] = age_purch_df["Average Purchase Price"].map("${:,.2f}".format)
age_purch_df["Total Purchase Value"] = age_purch_df["Total Purchase Value"].map("${:,.2f}".format)
age_purch_df["Avg. Total Purchase per Person"] = age_purch_df["Avg. Total Purchase per Person"].map("${:,.2f}".format)
#Create dataframe
age_purch_df = pd.DataFrame(age_purch_df)
#print (prints ok with format experiment. order of format/dataframe creation may matter in other examples)
age_purch_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Groupby SN to group transactions with player
Spenders = purchase_data.groupby("SN")
#Create dictionary of lists with desired values
Spenders_df = {"Purchase Count": Spenders["Purchase ID"].count(),
               "Average Purchase Price": Spenders["Price"].mean(), 
                "Total Purchase Value": Spenders["Price"].sum()}
#Create the data frame
Spenders_df = pd.DataFrame(Spenders_df)

#Sort dataframe by Total Purchase Value (Descending)
Top_Spenders_df = Spenders_df.sort_values("Total Purchase Value", ascending=False)

#Format the data
Top_Spenders_df["Average Purchase Price"] = Top_Spenders_df["Average Purchase Price"].map("${:,.2f}".format)
Top_Spenders_df["Total Purchase Value"] = Top_Spenders_df["Total Purchase Value"].map("${:,.2f}".format)

#Display the frame
Top_Spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve desired columns
Items = purchase_data[["Item ID", "Item Name", "Price"]]

#GroupBy Item ID and Item Name
Items = Items.groupby(["Item ID", "Item Name"])
#Create dictionary of lists with desired values
Items_df = {"Purchase Count": Items["Item ID"].count(),
           "Average Item Price": Items["Price"].sum()/Items["Item ID"].count(),
           "Total Purchase Value": Items["Price"].sum()}

#Create frame for results
Items_df = pd.DataFrame(Items_df)
#Sort in descending order by Purchase Count
Items_df_sort = Items_df.sort_values("Purchase Count", ascending=False)

#format
Items_df_sort["Average Item Price"] = Items_df_sort["Average Item Price"].map("${:,.2f}".format)
Items_df_sort["Total Purchase Value"] = Items_df_sort["Total Purchase Value"].map("${:,.2f}".format)
#display
Items_df_sort.head()
```
